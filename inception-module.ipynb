```python
from keras.layers import Input, Conv2D, MaxPooling2D, concatenate

def inception_module(x):
    # 1x1 convolution
    conv1x1_1 = Conv3D(64, (1, 1, 192), strides=1, padding='same',  activation='relu')(x)

    # 1x1 convolution followed by 3x3 convolution
    conv1x1_2 = Conv3D(96, (1, 1, 192), strides=1, strides=1, padding='same', activation='relu')(x)
    conv3x3 = Conv3D(128, (3, 3, 192), strides=1, padding='same', activation='relu')(conv1x1_2)

    # 1x1 convolution followed by 5x5 convolution
    conv1x1_3 = Conv3D(16, (1, 1, 192), strides=1, padding='same', activation='relu')(x)
    conv5x5 = Conv3D(32, (5, 5, 192), strides=1, padding='same', activation='relu')(conv1x1_3)

    # 3x3 max pooling followed by 1x1 convolution
    maxpool = MaxPooling3D((3, 3, 192), strides=1, padding='same')(x)
    conv1x1_4 = Conv3D(32, (1, 1), padding='same', activation='relu')(maxpool)

    # Concatenate the outputs along the channel axis
    inception_block = concatenate([conv1x1_1, conv3x3, conv5x5, conv1x1_4], axis=-1)

    return inception_block
```
